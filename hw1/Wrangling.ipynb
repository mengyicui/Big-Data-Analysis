# Step 5.1

```python
p = pd.read_sql_query('SELECT * FROM performance_df ORDER BY RANDOM() LIMIT 1000', con=engine)
r = pd.read_sql_query('SELECT * FROM routes_df', con=engine)
r
```

```python
import numpy as np
p = p.select_dtypes(exclude = [np.number])
r = r.select_dtypes(exclude = [np.number])

Jaccard1 = {}
Jaccard2 = {}
def Jaccard_distance(df1, df2):
    return len(set(df1).intersection(set(df2)))/len(set(df1).union(set(df2)))


for col_r in r:
    for col_p in p:
        result = Jaccard_distance(set(p[col_p]), set(r[col_r]))
        if result > 0:
            Jaccard1[col_p] = result
        Jaccard2[col_r] = Jaccard1
    Jaccard1= {}
Jaccard2
```

# Step 5.2.1

```python
p = pd.read_sql_query('SELECT * FROM performance_df ORDER BY RANDOM() LIMIT 1000', con=engine)
r = pd.read_sql_query('SELECT * FROM routes_df', con=engine)
combine = pd.DataFrame(r.merge(p, left_on=['destination_airport_code','itat_or_icao_code','source_airport_code'],right_on=['Dest','Carrier','Origin']))
combine
```

```python
pd.DataFrame(combine.groupby(['itat_or_icao_code','destination_airport_code'])['airline_id'].count()).reset_index().rename(columns={'airline_id':'count'})
```

# Step 5.2.2

```python
pd.DataFrame(combine[(combine['ArrDelayMinutes']>=30)].groupby(['itat_or_icao_code','destination_airport_code'])['airline_id'].count()).reset_index().rename(columns={'airline_id':'count'})
```

# Step 5.2.3

```python
re = pd.DataFrame(combine[(combine['ArrDelayMinutes']>=30)].groupby(['itat_or_icao_code','destination_airport_code'])['airline_id'].count()).reset_index().rename(columns={'airline_id':'count'})
re = pd.DataFrame(re[(re['count']>1)])
re
```
