```python
from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np
# data = ndimage.imread('data-behind-everything.jpg')
# plt.imshow(data)
```

```python
def convert_to_grayscale(image, crop_left, crop_top, crop_right, crop_bottom, contrast_scale):
    crop_image = image[crop_left:-crop_right, crop_top:-crop_bottom]
    gray_data = np.mean(crop_image.transpose(1,2,0),axis = 1).transpose()
    grayscale_median = np.median(crop_image)
    for x in range(0, gray_data.shape[0]):
        for y in range(0, gray_data.shape[1]):
            n = gray_data[x, y]
            if n < grayscale_median:
                gray_data[x, y] = 0.7 * n
                if gray_data[x, y] < 0:
                    gray_data[x, y] = 0
            elif n > grayscale_median:
                gray_data[x, y] = 1.3 * n
                if gray_data[x, y] > 255:
                    gray_data[x, y] = 255
            else:
                gray_data[x, y] = n
    return gray_data
```

```python
data = ndimage.imread('data-behind-everything.jpg')
```

```python
gray_data = convert_to_grayscale(data, 10, 50, 10, 50, 0.3)
```
